Z-puan, veri setindeki bir gzlemin ortalamadan ka standart sapma uzaklkta olduunu gsterir.
Pozitif z-skoru, gzlemin ortalamadan byk olduunu,
Negatif z-skoru, gzlemin ortalamadan kk olduunu,

Herhangi bir normal dalm, verilerin standartlatrlmas ile z-dalmna dntrlebilir. Verinin standartlatrlmas, hem farkl standart sapma ve ortalamaya sahip verilerin karlatrlmasn kolaylatrr hem de kolaylkla olaslk hesab yaplmasna imkan salar. ## 1. Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
# scaled edilen X ile devam ediyoruz.
lm.fit(X_train_scaled, y_train)
```

```python
y_pred = lm.predict(X_test_scaled) # y_pred alrken yine scaled edilmi test ve train datalarn alyoruz
y_train_pred = lm.predict(X_train_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_val(y_train, y_train_pred, y_test, y_pred, i):
    
    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
              
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    return pd.DataFrame(scores)
```

```python
ls =train_val(y_train, y_train_pred, y_test, y_pred, "linear")       # modelin ad
ls
```

* R2_scored deerleri aras baya alm
* mae ve mse de artlar belirgin

```python
# function for set text color of positive
# values in Dataframes
# korelasyona gre renklendirerek daha net grebiliriz.

def color_red(val):
    
    if (0.90 < val < 1) or (-1 < val<-0.90):
        color = 'red'
    elif val == 1 :
        color = 'blue'
    else:
        color = 'black'
    return f'color: {color}'
```

## 5. Grid Search (identify the optimal hyperparameters)

```python
from sklearn.model_selection import GridSearchCV
```

```python
elastic_model = ElasticNet(max_iter =10000, random_state=42)
```

```python
param_grid = {"alpha":[0.01,0.012,0.2,0.5,0.6,0.7,1],
             "l1_ratio":[0.1,0.5,0.7,0.9,0.95,0.99,1]}
```

```python
grid_model = GridSearchCV(estimator = elastic_model, param_grid = param_grid, scoring = 'neg_root_mean_squared_error',
                         cv =5, verbose =2)
```

```python
grid_model.fit(X_train_scaled, y_train)
```

```python
grid_model.best_params_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
grid_model.best_index_
```

```python
grid_model.best_score_
```
