## <span style=" color:#bf2e98">Introduction

* It works by creating a number of Decision Trees during the training phase. Each tree is constructed using a **random subset of the data set to measure a random subset of features in each partition**. This randomness introduces variability among individual trees, **reducing the risk of overfitting and improving overall prediction performance**.
* In prediction, the algorithm aggregates the results of all trees, either by **voting** (for classification tasks) or by averaging (for regression tasks).



#### What are Ensemble Learning models?
*  In ensemble learning, different models team up to enhance predictive performance - think of them as a bunch of friends with different strengths tackling a problem together. Its all about leveraging the collective **wisdom of the group** to overcome individual limitations and make more informed decisions in various machine learning tasks.
* Some popular ensemble models include- **XGBoost, AdaBoost, LightGBM, Random Forest, Bagging, Voting etc**.
#### What is Bagging and Boosting?
* **Bagging** is an ensemble learning model, where multiple week models are trained on different subsets of the training data. Each subset is sampled with replacement and prediction is made by **averaging the prediction** of the week models **for regression** problem and considering **majority vote for classification problem**.
* **Boosting** trains multiple based models sequentially. In this method, each model tries to correct the errors made by the previous models. Each model is trained on a modified version of the dataset, the instances that were misclassified by the previous models are given more weight. The final prediction is made by **weighted voting**.

## Train | Test Split

```python
# Drop the target column (species) during get_dummies and assign thre rest to X
X = pd.get_dummies(df.drop('species',axis=1),dtype=int, drop_first=True) 

# Target column
y = df['species']
```

## <span style=" color:#eb346e">Random Forest Hyperparameters

### 1. Increasing the Predictive Power
**n_estimators:** It is the number of trees the algorithm builds before taking the maximum voting or taking the averages of predictions. In general, **a higher number of trees increases the performance and makes the predictions more stable**, but it also slows down the computation.

**max_features:** The maximum number of features random forest considers to split a node. Sklearn provides several options.

**min_sample_leaf:** This determines the minimum number of leafs required to split an internal node.

#### How to choose number of trees?
* Reasonable default value: 100
* Publications suggest **64-128** trees.
* Cross-Validate a **grid search** of trees.
* **Plot** Error versus number of trees (similar to elbow method of KNN).
* Should notice diminishing error reduction after some N trees.
* After a certain number of trees, these **issues** can occur:
  * Different random selections do not reveal any more information.
  * Trees become highly correlated.
  * Different random selections are simply duplicating trees that have already been created.
#### How many features to include in each subset when splitting at a node?
* In original publicaiton: Suggested subset of **log2(N+1)** random features in subset given a set of N total features.
* Current suggested convension is **sqrt(N)** in the subset given N features.
* Suggestions by Breiman indicated **N/3** may be more suitable **for regression tasks**, typically larger than sqrt(N).

### 2. Increasing the Random Forest Models Speed
**n_jobs:** It tells the engine how many processors it is allowed to use. If it has a value of one, it can only use one processor. A value of -1 means that there is no limit.

**oob_score:** Also called **oob sampling**, which is a random forest **cross-validation method**. In this sampling, about one-third of the data is not used to train the model and can be used to evaluate its performance. These samples are called the **out-of-bag samples**. Its very similar to the leave-one-out-cross-validation method, but almost no additional computational burden goes along with it.
</span>

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# See the parameters:
# help(RandomForestClassifier)
```

```python
# Use 10 random trees
model = RandomForestClassifier(n_estimators=10,random_state=101) 
```

```python
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

## Evaluation

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

# To plot the confusion matrix

from sklearn.metrics import ConfusionMatrixDisplay
```

```python
confusion_matrix(y_test,preds)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, preds);
```

There are 2 incorrect predictions. In general, ML classification models have trouble predicting the Chinstrap category in the Penguins dataset. However, the RF model is quite good at predicting it.

```python
print(classification_report(y_test, preds))
```

It performs better than the Decision Tree model that we used in other notebook.

## Choosing correct number of trees

Let's explore if continually adding more trees improves performance...

```python
test_error = []

for n in range(1,40):
    # Use n random trees
    model = RandomForestClassifier(n_estimators=n, random_state=101)
    model.fit(X_train,y_train)
    test_preds = model.predict(X_test)
    test_error.append(1-accuracy_score(test_preds,y_test))
```

```python
plt.plot(range(1,40),test_error,label='Test Error')
plt.legend()
```

<div class="alert alert-warning alert-info">

<span style=" color:#bf2e98">
We can see that there is not much change after 3-5 trees. Since the dataset is small, consisting of 6 features and a target variable, it provides enough information. Compared to the Decision Trees model used in the previous notebook, Random Forest gives higher scores on the Chinstrap category and overall accuracy.
